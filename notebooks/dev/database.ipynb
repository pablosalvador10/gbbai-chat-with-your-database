```python
def execute_query(connection, query):
    # Create a cursor object
    cursor = connection.cursor()

    # Execute the query
    cursor.execute(query)

    # Fetch all the rows
    rows = cursor.fetchall()

    # Return the rows
    return rows
```

```python
# Define your SQL query
query = "SELECT * FROM expanded_detroit_tigers_baseball_stats"  # replace with your table name

# Execute the query and get the result
result = execute_query(conn, query)

# Print the result
for row in result:
    print(row)
```

```python
import json

# Open the JSON file
with open("tests\\validation\\querys.jsonl", "r") as file:
    # Load the JSON object from the file
    data = json.load(file)

    # Iterate over each query in the queries list
    for query in data["queries"]:
        # Extract the sql_query and natural_language_query from the query
        sql_query = query["sql_query"]
        natural_language_query = query["natural_language_query"]

        # Execute the sql_query and get the result
        result = execute_query(conn, sql_query)

        # Convert the result to a list
        result_list = [row[0] for row in result]

        # Print the natural_language_query and sql_query
        print(f"Natural Language Query: {natural_language_query}")
        print(f"SQL Query: {sql_query}")

        # Check if the result matches the one in the JSON file
        if result_list == query["results"]:
            print("The results match.")
        else:
            print("The results do not match.")

        # Print a separator for readability
        print("-" * 60)
```
